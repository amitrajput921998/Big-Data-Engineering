Pandas is apowerful data manipulation library in python, widely used for dtaa analysis and data cleaning.It provides two primary data structure.
Series and DataFrame. A Series is a one-dimensional array-like object containing an array of data and an associated array of data labels.DataFrame is a two-dimensional table-like data structure with rows and columns.


```python
import pandas as pd
```

```python
data=[1,2,3,4,5,6]
series=pd.Series(data)
print(series)
```

```python
data={'a':1,'b':2,'c':3,'d':4,'e':5}
series=pd.Series(data)
print(series)
```

```python
data=[10,20,30]
index=['a','b','c']
series=pd.Series(data,index=index)
print(series)
```
